# File preprocessing notebook
This notebook aim to apply some cleaning and filters for future data analysis operations

## First step : Importation of pandas library and creation of the dataframe

```python
import pandas as pd
```

```python
df = pd.read_csv('mapstr.csv')
```

```python
df
```

## Second step : Filtering out the Private addresses, replacing the NaN values for comments to empty string and desc ordering to have the oldests addresses first 

```python
privateless_df = df[~df['tags'].str.contains("Private")]
```

```python
privateless_df['userComment'] = privateless_df['userComment'].fillna("")
```

```python
privateless_df
```

```python
privateless_df = privateless_df.iloc[::-1]
```

```python
privateless_df = privateless_df.reset_index(drop=True)
```

```python
privateless_df
```

# Tags separations

## Columns split based on the # character

```python
df_tags_columns = privateless_df['tags'].str.split('#', expand=True)
```

## Columns renamed based on the max number of tags

```python
df_tags_columns.columns = [f'tag{i+1}' for i in range(df_tags_columns.shape[1])]
```

```python
df_tags_columns
```

## Concatenation of the tags columns with the original dataframe

```python
privateless_df = pd.concat([privateless_df, df_tags_columns], axis=1)
```

```python
privateless_df = privateless_df.drop("tags", axis=1)
```

## Address separation

```python
adr_df = privateless_df['address'].str.split(',', expand=True)
```

```python
adr_df.columns = [f'adr{i+1}' for i in range(adr_df.shape[1])]
```

## Postal Code Separation

```python
df_pc = adr_df["adr2"].str.split(' ', expand=True)
```

```python
df_pc.columns = [f'cp{i+1}' for i in range(df_pc.shape[1])]
```

```python
df_pc = df_pc[["cp2","cp3"]]
```

```python
print(privateless_df.shape,adr_df.shape,df_pc.shape)
```

## Address concatenation

```python
df = pd.concat([privateless_df,adr_df, df_pc], axis=1)
```

```python
df
```

## Removal of unused columns

```python
df = df.drop(columns=['address','adr2','adr4','adr5','adr6'],axis=1)
```

## Renaming of columns for better readability

```python
df = df.rename(columns={'adr1':'address', 'adr3':'country', 'cp2':'postalCode', 'cp3':'city'})
```

## Reordering of the columns

```python
df = df[['name','address', 'city', 'postalCode', 'country', 'icon', 'userComment', 'tag1', 'tag2', 'tag3', 'tag4', 'tag5', 'tag6', 'tag7']]
```

## Replacement of NaN and None values with ""

```python
df = df.fillna("")
```
